```python
# Also bin needs to be included in the groupby
crsp['bin_filled'] = crsp.groupby(['permno'])[['bin']].ffill()
crsp['pweight_filled'] = crsp.groupby(['permno','bin_filled'])[['pweight']].ffill()
```

```python
crsp = crsp.sort_values(['permno', 'yyyymm', 'bin'], ascending=[True, True, True])
```

```python
crsp.groupby('permno')
```

```python
# Drop NaN in pweight filled, they are related to stocks that started 
# trading before the rebalancing period, before being included 
# in the rebalanced portfolio

crsp.dropna(subset=['pweight_filled'], inplace=True)
```

```python
portfolio_returns = crsp.groupby(['yyyymm','bin_filled']).apply(lambda x: sum(x['ret']*x['pweight_filled'])).reset_index(name='portfolio_ret')
```

```python
portfolio_returns['bin_filled'] = portfolio_returns.bin_filled.astype(int)
```

```python
portfolio_returns
```

```python
portfolio_returns = portfolio_returns.pivot(index='yyyymm',columns='bin_filled',values='portfolio_ret').reset_index()
```

```python
portfolio_returns = portfolio_returns.rename_axis(None, axis=1)
```

```python
portfolio_returns
```

```python
# Calculating long-short returns for quantiles
portfolio_returns[(str(portfolio_returns.columns[-1])+'-'+str(portfolio_returns.columns[1]))] = portfolio_returns.iloc[:,-1] - portfolio_returns.iloc[:,1]
```

# Calculate t-stats and Information ratio

### Information ratio compared to MOM

```python
FFMom = pd.read_csv(dataPath+'/F-F_Momentum_Factor.CSV', skiprows=13, index_col=0, skipfooter=100, engine='python')
FFMom.rename(columns={'Mom   ': 'Mom'}, inplace=True)
```

```python
FFMom = FFMom.reset_index().rename(columns={'index':'yyyymm'})
```

```python
portfolio_returns = portfolio_returns.merge(FFMom, on=['yyyymm'], how='left')
```

```python
portfolio_returns['Mom   ']
```

```python
# The tracking error is defined as the standard deviation of the difference between
# the strategy return minus the benchmark return
portfolio_returns['diff'] = (portfolio_returns['2.0-0.0'] - portfolio_returns['Mom   '])
```

##### I will try by annualizing returns

```python
portfolio_returns['year'] = (portfolio_returns['yyyymm']//100)
```

```python
portfolio_returns['Mom_ret'] = (portfolio_returns['Mom']/100+1)
portfolio_returns['L-S_ret'] = (portfolio_returns['2.0-0.0']+1)
```

```python
portfolio_returns['Mom_cum_return'] = (portfolio_returns.groupby('year')['Mom_ret'].cumprod())-1
portfolio_returns['Mom_cum_return'] = (portfolio_returns.groupby('year')['L-S_ret'].cumprod())-1
```

```python
yearly_port_ret = portfolio_returns.groupby('year').tail(1)
```

```python
yearly_port_ret = yearly_port_ret[['year', 'L-S_ret', 'Mom_cum_return']]
```

```python
yearly_port_ret['diff'] = (yearly_port_ret['L-S_ret'] - yearly_port_ret['Mom_cum_return'])
```

```python
import numpy as np
```

### Calculation of Information Ratio using yearly cumulated returns

```python
(yearly_port_ret['L-S_ret'].mean() - yearly_port_ret['Mom_cum_return'].mean())/np.std(yearly_port_ret['diff'])
```
